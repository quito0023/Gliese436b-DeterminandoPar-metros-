```python
from sympy import *
init_printing(pretty_print=true)
import math
π = math.pi
x = symbols('x')
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Dados da Estrela
M_sS = 0.4411671425335 # Massa de GJ436 em M⨀
R_sS = 0.425431996583939 # Raio de GJ436 em R⨀
Ts = 3586.106561295 # Temperatura de GJ 436 em Kelvin
#_________________________________________________

# Dados do Planeta
P_d = 2.64388561055203 # Período orbital de GJ436-b em dias
e = 0.145 # Excentricidade da órbita de GJ436-b
Vrad = 17.12 # Velocidade Radial de GJ436-b em m/s 
Tdepth = 0.6819 # Profundidade de trânsito de GJ436-b
inc = 86.858 # Inclinação da órbita de GJ436-b em graus 
#_________________________________________________

# cte´s em S.I
σ = 5.67*10**(-8) # Constante de stephanboltzmann
G = 6.67*10**(-11) # Constante gravitacional
A = 0.367 # Albedo da Terra
MS = 1.98892*10**(30) # Massa do Sol
MJ = 1.8987*10**(27) # Massa de Júpiter
MT = 5.9742*10**(24) # Massa da Terra
RS = 696340000 # Raio do Sol
RJ = 69911000 # Raio de Júpiter
RT = 6378100 # Raio da Terra
TS = 5778 # Temperatura do Sol
LS = 3.87*10**(26) # Luminosidade do Sol
Ua = 1.496*10**(11) # Unidade Astronômica
#_________________________________________________
```

```python
# Convertendo dados pra S.I
Ms = M_sS*MS # Massa de GJ436 em kilogramas
Rs = R_sS*RS # Raio de GJ436 em metros
Pp = P_d*86400 # Período orbital de GJ436-b em segundos
i = (inc*π)/180
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("Massa da estrela em S.I:")
print(Ms,"kg")
print("_________________________________________________")
print("Raio da estrela em S.I:")
print(Rs,"m")
print("_________________________________________________")
print("Período orbital do exoplaneta em S.I:")
print(Pp,"s")
print("_________________________________________________")
print("Inclinação do exoplaneta em radianos:")
print(i,"rad")
print("_________________________________________________")
#_________________________________________________
```

```python
# Calculando infos. da estrela
print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("              Luminosidade de GJ436               ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")

Ls = 4*π*σ*(Rs**2)*(Ts**4) # Luminosidade de GJ436
L_sS = (Ls/LS) # Luminosidade de GJ436 em L⨀
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("A luminosidade da estrela em Watt:")
print(Ls,"W")
print("A luminosidade da estrela em luminosidades solares:")
print(L_sS,"L⨀")
print("_________________________________________________")

#_________________________________________________
print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("            Zona Habitável de GJ436               ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")

Z_Hi = 0.75*(Ls/LS)**(1/2) # Zona Habitável interna de GJ436
Z_Hc = 1*(Ls/LS)**(1/2) # Zona Habitável central de GJ436
Z_Hf = 1.77*(Ls/LS)**(1/2) # Zona Habitável externa de GJ436
d_Hi = (((Ls)/(4*π*σ*((373)**(4))))**(1/2))/(Ua) # Distância até o inicio da Zona Habitável de GJ436
d_Hf = (((Ls)/(4*π*σ*((273)**(4))))**(1/2))/(Ua) # Distância até o final da Zona Habitável de GJ436
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("A Zona Habitável da estrela vai de:")
print("Zona interna:",Z_Hi,"U.A ;Zona central:",Z_Hc,"U.A ;Zona final:",Z_Hf,"U.A ")
print("_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _")
print("Distância até a Zona Habitável de forma menos criterioza incial:",d_Hi,"U.A")
print("Distância até a Zona Habitável de forma menos criterioza final:",d_Hf,"U.A")
print("_________________________________________________")

#_________________________________________________
print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("           Baricentro do Sistema GJ436              ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")

 # Cálculos de baricentro através da análise da velocidade radial de GJ436-b
B_s_a = ((Vrad*Pp*((1-(e**(2))**(1/2))))/(2*π)) # Distância de GJ436 até o baricentro, em metros
Bs_Ua_a = (B_s_a)/Ua # Distância de GJ436 até o baricentro, em U.A
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("Semieixo-maior da órbita de GJ436 em torno do centro de massa, em metros:")
print(B_s_a,"m")
print("Em U.A.:",Bs_Ua_a,"U.A")
```

```python
# Calculando infos. do exoplaneta
print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("       Semieixo-maior da órbita de GJ436-b        ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
a = ((G*Ms*(Pp**2))/(4*(π**2)))**(1/3) # Semieixo-maior de GJ436-b em metros
a_Ua = a/Ua # Semieixo-maior de GJ436-b em U.A
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("O semieixo-maior da órbita do planeta em metros:")
print(a,"m")
print("Em U.A.:",a_Ua,"U.A")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("           Raio do exoplaneta GJ436-b             ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
Rp = ((Rs)*((((Tdepth/100)))**(1/2))) # Raio de GJ436-b em metros
R_pT = Rp/RT # Raio de GJ436-b em Raio de Júpiter
R_pJ = Rp/RJ # Raio de GJ436-b em Raio da Terra
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("O raio do planeta em metros:")
print(Rp,"m")
print("Em função de raios Terrestre:",R_pT,"R⨁")
print("Em função de raios de Jupiter:",R_pJ,"RJ")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("       Tempo de trânsito do exoplaneta GJ436-b      ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
Pt = ((4*a*(((Rs+Rp)**(2))-(a*cos(i))**(2)))/(G*Ms))**(1/2) # Tempo de trânsito em segundo 
Pt_h = Pt/3600 # Tempo de trânsito em horas
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("O período de trânsito por uma sessão do diâmetro de sua estrela em segundos:")
print(Pt,"s")
print("Em horas:",Pt_h,"h")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("     Velocidade orbital do exoplaneta GJ436-b       ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
vp_c = (2*π*a)/(Pp) # Velocidade em que GJ436-b orbita sua estrela, considerando uma órbita circular, em m/s
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
print("Velocidade considerando movimento circular:",vp_c,"m/s")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("            Massa do exoplaneta GJ436-b             ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
Mp_c = ((Vrad*((Pp)**(1/3))*((1-e**(2))**(1/2))*((Ms)**(2/3)))/((sin(i))*((2*π*G)**(1/3)))) # Massa de GJ436-b, utilizando velocidade radial, em kg 
Mpc_J = ((Mp_c)/(MJ)) # Massa de GJ436-b, utilizando velocidade radial, em MJ 
Mpc_T = ((Mp_c)/(MT)) # Massa de GJ436-b, utilizando velocidade radial, em M⨁ 
print("Massa, utilizando a velocidade na órbita circular, em metros:")
print(Mp_c,"kg")
print("Em função de Massas Terrestre:",Mpc_T,"M⨁")
print("Em função de Massas de Jupiter:",Mpc_J,"MJ")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("          Densidade do exoplaneta GJ436-b           ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
V = (4*π*(Rp**3))/(3) # Volume de GJ436-b
ρ_c = (Mp_c)/(V) # Densidade de GJ436-b, utilazando massa calculada com a velocidade circular, em kg/m³
ρc = ρ_c/1000 # Densidade de GJ436-b, utilazando massa calculada com a velocidade circular, em g/cm³
print("Densidade, utilazando a massa calculada com a velocidade circular: ")
print(ρc,"g/cm³")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("Temperatura de equilíbrio e Insolação do exoplaneta GJ436-b")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
Tp_eq = Ts*(((Rs)/(2*a))**(1/2))*((1-A)**(1/4)) # Temperatura de GJ436-b, supondo um corpo negro (emissor e absorssor perfeito), em Kelvin 
Fp = Ls/(a**2) # Fluxo de GJ436 que chega até a distância da órbita e GJ436-b, em Watts/m³
Insol = (Fp)/(LS/(Ua**2)) # Fluxo de Insolação recebido por GJ436-b; Fluxo recebido por GJ436-b de GJ436, em relação ao fluxo que a Terra recebe pelo Sol
print("Temperatura de equlíbrio do exoplaneta, em Kelvin:")
print(Tp_eq,"K")
print("_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _")
print("Fluxo de energia recebido pelo exoplaneta por sua estrela, em W/m³:")
print(Fp,"W/m³")
print("Fluxo em relação ao fluxo recebido da Terra pelo Sol:",Insol,"F⨁")
print("_________________________________________________")

#_________________________________________________

print("◤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◥")
print("        Probabilidade de Trânsito de GJ436-b        ")
print("◣▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤▤◢")
#_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
Prob = ((Rs+Rp)/(a*(1-(e**2))))*100
print("Probabilidade do trânsito do exoplaneta, em porcentagem:")
print(Prob,"%")
```
